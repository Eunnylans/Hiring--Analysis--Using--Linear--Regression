## Observations:

### Correlation Coefficients Explanation:

- **1 indicates a perfect positive correlation**: This means that as one variable increases, the other variable also increases in a perfectly linear relationship. For example, a correlation coefficient of **1** between two variables would mean that for every unit increase in one variable, the other variable increases by a consistent amount.

- **-1 indicates a perfect negative correlation**: This means that as one variable increases, the other variable decreases in a perfectly linear relationship. For example, a correlation coefficient of **-1** would mean that for every unit increase in one variable, the other variable decreases by a consistent amount.

- **0 indicates no correlation**: This means there is no linear relationship between the two variables; changes in one variable do not predict changes in the other.
